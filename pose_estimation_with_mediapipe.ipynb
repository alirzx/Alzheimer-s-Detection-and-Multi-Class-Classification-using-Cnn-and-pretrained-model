```python
def calculate_angle (a,b,c) :
  a = np.array(a) #first
  b = np.array(b) # Mid
  c = np.array(c) # End

  radians = np.arctan2(c[1]-b[1], c[0]-b[0])  - np.arctan2(a[1]-b[1], a[0]-b[0])
  angle = np.abs(radians*180.0/np.pi)

  if angle > 180.0 :
    angle= 360 - angle

  return angle
```

```python
import cv2
import mediapipe as mp
import numpy as np

mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

# Initialize MediaPipe Pose model
pose = mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5, min_tracking_confidence=0.5)

# Open video stream
cap = cv2.VideoCapture(0)

while True:
    # Read a frame from the video stream
    ret, image = cap.read()

    if not ret:
        break

    # Convert image to RGB format
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Process image with MediaPipe Pose model
    results = pose.process(image)

    # Convert image to BGR format
    #image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

    # Check if pose landmarks are detected
    if results.pose_landmarks is not None:

        # Extract left elbow coordinates
        left_elbow_x = results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].x * image.shape[1]
        left_elbow_y = results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].y * image.shape[0]

        # Extract left shoulder coordinates
        left_shoulder_x = results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].x * image.shape[1]
        left_shoulder_y = results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].y * image.shape[0]

        # Calculate direction vector of shoulder-elbow line
        shoulder_elbow_direction = np.array([left_shoulder_x - left_elbow_x, left_shoulder_y - left_elbow_y])

        # Calculate ending point of line based on direction of shoulder-elbow line
        line_length = 100 # Change this value to adjust length of line
        line_ending_point = np.array([left_elbow_x, left_elbow_y]) + line_length * shoulder_elbow_direction / np.linalg.norm(shoulder_elbow_direction)

        # Draw line from elbow to ending point
        line_thickness = 10 # Change this value to adjust line thickness
        line_color = (0, 0, 255) # Change this value to adjust line color
        cv2.line(image, (int(left_elbow_x), int(left_elbow_y)), (int(line_ending_point[0]), int(line_ending_point[1])), line_color, line_thickness)

        # Display pose landmarks on image
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

    # Display image with line
    cv2.imshow("Video Stream with Line", image)

    # Exit on key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    #test

# Release resources
cap.release()
cv2.destroyAllWindows()
```
